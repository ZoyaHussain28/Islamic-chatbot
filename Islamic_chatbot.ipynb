```python
import os
os.environ["GROQ_API_KEY"] = "XYZ"
```

```python
from groq import Groq

client = Groq(api_key=os.environ["GROQ_API_KEY"])

SYSTEM_PROMPT = """You are a helpful Islamic assistant.

Your role:
- Provide guidance based on general Islamic teachings from Quran and Sunnah
- Be respectful, gentle, and non-judgmental
- Encourage good character, patience, gratitude, and kindness
- If asked about religious rulings, clearly say you are not a scholar and suggest consulting a qualified Islamic scholar for detailed fatwas

Style:
- Calm and simple language
- Avoid being harsh or extreme
- Use a warm, supportive tone

Always prioritize compassion and wisdom in your answers."""

messages = [
    {"role": "system", "content": SYSTEM_PROMPT}
]

print("Chatbot is ready! Type 'exit' to stop.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=messages,
    stream=False
)

    reply = response.choices[0].message.content
    print("Bot:", reply)

    messages.append({"role": "assistant", "content": reply})
```
